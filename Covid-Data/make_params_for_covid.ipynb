```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import shapely
import seaborn as sns
```

```python
# Modify COVID dataframes and added more parameters to it - 
# Difference: Increase/decrease in new infections from the day before within the state/province
```

```python
atlCases = pd.read_csv('atlCases.csv')
jfkCases = pd.read_csv('jfkCases.csv')
lgaCases = pd.read_csv('lgaCases.csv')
seaCases = pd.read_csv('seaCases.csv')
laxCases = pd.read_csv('laxCases.csv')
yvrCases = pd.read_csv('yvrCases.csv')
yegCases = pd.read_csv('yegCases.csv')
yyzCases = pd.read_csv('yyzCases.csv')
yulCases = pd.read_csv('yulCases.csv')
```

```python
temp = yegCases.shift(1)
yegCases['Difference'] = yegCases['Confirmed'] - temp['Confirmed']
#yegCases = yegCases.drop(columns = ['Count'])
yegCases
```

```python
temp = yvrCases.shift(1)
yvrCases['Difference'] = yvrCases['Confirmed'] - temp['Confirmed']
#yvrCases = yvrCases.drop(columns = ['Count'])
yvrCases
```

```python
temp = yulCases.shift(1)
yulCases['Difference'] = yulCases['Confirmed'] - temp['Confirmed']
#yulCases = yulCases.drop(columns = ['Count'])
yulCases
```

```python
temp = yyzCases.shift(1)
yyzCases['Difference'] = yyzCases['Confirmed'] - temp['Confirmed']
#yyzCases = yyzCases.drop(columns = ['Count'])
yyzCases
```

```python
# For American Count/Difference - 
# The 'Difference' column is aggregated by counties and not states

lgaCases = lgaCases.sort_values(by = ['County', 'Date'])
temp = lgaCases.shift(1)
lgaCases['Difference'] = lgaCases['Confirmed'] - temp['Confirmed']
lgaCases
```

```python
jfkCases = jfkCases.sort_values(by = ['County', 'Date'])
temp = jfkCases.shift(1)
jfkCases['Difference'] = jfkCases['Confirmed'] - temp['Confirmed']
jfkCases
```

```python
atlCases = atlCases.sort_values(by = ['County', 'Date'])
temp = atlCases.shift(1)
atlCases['Difference'] = atlCases['Confirmed'] - temp['Confirmed']
atlCases
```

```python
laxCases = laxCases.sort_values(by = ['County', 'Date'])
temp = laxCases.shift(1)
laxCases['Difference'] = laxCases['Confirmed'] - temp['Confirmed']
laxCases
```

```python
seaCases = seaCases.sort_values(by = ['County', 'Date'])
temp = seaCases.shift(1)
seaCases['Difference'] = seaCases['Confirmed'] - temp['Confirmed']
seaCases = seaCases.drop(columns = ['Count'])
seaCases
```
